## Modelos

Vamos a crear el primer modelo para poder mapear a la base de datos. Para esto es necesario que creemos una clase que herede de la clase Base de SQLAlchemy

```python
from sqlalchemy.orm import declarative_base
from sqlalchemy import Column, Integer, String

#Clase base para heredar y construir modelos.
Base = declarative_base()

#Clase modelo para mapear una tabla.
class Perro(Base):
  __tablename__="perrotest"

  id = Column(Integer, primary_key=True)
  name = Column(String,)

  def __repr__(self) -> str:
      return super().__repr__()
```

```python
Perro.__table__
```

##Crear tabla
Ejecutamos la siguiente linea de código para indicarle al engine de SQLAlchemy que debe crear la tabla perro con los atributos definidos en la clase anterior.

```python
Base.metadata.create_all(engine)
```

## Ahora vamos a insertar datos.

Para insertar datos usamos la clase que creamos.

```python
perro1 = Perro(name = "lulu")
```

En este momento tenemos la instancia de la clase creada y la tabla creada en base de datos, sin embargo, no se ha realizado el INSERT en la base de datos con los datos del perro1. Para esto es necesario crear una sección e indicar que guarde los cambios como se muestra en el siguiente código:


```python
from sqlalchemy.orm import sessionmaker
#Creamos la sección
Session = sessionmaker(bind=engine)
session = Session()

#Agregamos el objeto a la seccion
session.add(perro1)

#Realizamos el guardado en la base de datos.
session.commit()
```

Ahora vamos a consultar con ayuda de la sección el objeto que insertamos en base de datos como registro:


```python
for perro in session.query(Perro).filter_by(name='lulu').all():
  print(perro)
```
